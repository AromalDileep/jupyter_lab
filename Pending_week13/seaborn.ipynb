# Seaborn

```python
import seaborn as sns
```

```python
# Basic plotting

tips = sns.load_dataset('tips')
tips.head()
```

```python
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill', y ='tip', data = tips)
```

```python
# line plot
sns.lineplot(x='size', y ='total_bill', data=tips)
plt.title('Line plot of Total bill by size')
plt.show()
```

# Categorical plot

## barplot

```python
sns.barplot(x='day', y='total_bill', data = tips)
plt.title('Bar plot of total bill by day')
plt.show()
```

# boxplot

```python
sns.boxplot(x= 'day', y= 'total_bill',data= tips)
# can see percentile, outliers
```

# violinplot

```python
sns.violinplot(x='day', y='total_bill', data = tips)
# see mean
# see distribution of the data
```

# Histogram

```python
sns.histplot(tips['total_bill'], bins=10, kde= True)
# kde is kernal density estimator, smoothening the histogram, shows how the data is distributed
```

```python
# KDE plot
sns.kdeplot(tips['total_bill'], shade=True)
#used to visualize the probability density of a continuous variable
#Visualizing Data Distribution: 
```

```python
sns.pairplot(tips)
```
