```python
import seaborn as sns 
import matplotlib.pyplot as plt 

arr = [2, 3, 4, 2, 3, 6, 4, 2]

dt = {}

for i in arr:
    count = 1
    if i in dt:
        dt[i] += 1 
    else:
        dt[i] = count

key = []
value = []
for k, v in dt.items():
    key.append(k)
    value.append(v)

sns.barplot(x =key, y=value)


```
