## Feature Engineering

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.info()
```

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

## Drop Date of Journey

```python
df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x: x.split(' ')[0])
```

```python
df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_minute'] = df['Arrival_Time'].str.split(':').str[1]
df.head()
```

```python
df['Arrival_hour'] = df['Arrival_hour'].astype(int)
df['Arrival_minute'] = df['Arrival_minute'].astype(int)
```

```python
df.info()
```

```python
df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
df.info()
```

```python
df.head(2)
```

```python
df['Departure_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Departure_minute'] = df['Dep_Time'].str.split(':').str[1]
```

```python
df['Departure_hour'] = df['Departure_hour'].astype(int)
df['Departure_minute'] = df['Departure_minute'].astype(int)
```

```python
df.drop('Dep_Time',axis=1, inplace=True)
```

```python
df.info()
```

```python
df.head(2)
```

```python
df['Total_Stops'].unique()
```

```python
# map the nan value with no stop
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, '4 stops':4, np.nan:1})
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df.head(2)
```

```python
df.drop('Route', axis=1, inplace=True)
```

```python
df.info()
```

```python
df.head(2)
```

```python
df['Duration_Hour'] = (
    df['Duration']
    .str.split(' ')                # Split on space
    .str[0]                        # Extract the first part (e.g., '2h' or '5m')
    .str.replace('m', '', regex=False)  # Remove 'm' (if present)
    .str.replace('h', '', regex=False)  # Remove 'h'
    .fillna(0)                     # Replace missing values with 0
    .astype(int)                   # Convert to integer
)


# or
# df['Hours'] = df['Duration'].str.extract(r'(\d+)h')


# If you have a column 'Duration' with values like '2h 30m', '15h', '1h 45m', etc., 
# the regex r'(\d+)h' will match the number of hours (digits before 'h') and return them.
# 
```

```python
df.Duration_Hour
```

```python
df['Duration_minute'] =(
    df['Duration']
    .str.split(' ')
    .str[1]
    .str.split('m')
    .str[0].fillna(0)
    .astype(int)
 )
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
encoded = encoder.fit_transform(df[['Airline', 'Source', 'Destination']]).toarray()
```

```python
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
```

```python
df = pd.concat([df, encoded_df],axis=1)
```
