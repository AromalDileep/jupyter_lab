
1. Support Vector Classification (SVC) Hyperparameters:
- `C`: Regularization parameter (float, default=1.0)
  - Controls the trade-off between maximizing the margin and minimizing classification errors
  - Smaller values lead to a larger margin but more misclassifications
  - Larger values create a smaller margin with fewer misclassifications

- `kernel`: Specifies the kernel type (string, default='rbf')
  - 'linear': Linear kernel
  - 'poly': Polynomial kernel
  - 'rbf': Radial Basis Function (Gaussian) kernel
  - 'sigmoid': Sigmoid kernel

- `degree`: Degree of the polynomial kernel (int, default=3)
  - Only used when kernel is 'poly'

- `gamma`: Kernel coefficient (float, default='scale')
  - Defines how far the influence of a single training example reaches
  - 'scale': 1 / (n_features * X.var())
  - 'auto': 1 / n_features

- `coef0`: Independent term in kernel function (float, default=0.0)
  - Used in 'poly' and 'sigmoid' kernels

- `probability`: Whether to enable probability estimates (bool, default=False)

- `class_weight`: Weights associated with classes (dict or 'balanced')
  - Adjusts weights to handle imbalanced datasets

2. Decision Tree Hyperparameters:
- `criterion`: The function to measure the quality of a split (string, default='gini')
  - 'gini': Gini impurity
  - 'entropy': Information gain
  - 'log_loss': Cross-entropy

- `max_depth`: Maximum depth of the tree (int, default=None)
  - Controls tree complexity and prevents overfitting
  - None means nodes are expanded until all leaves are pure or contain fewer than min_samples_split samples

- `min_samples_split`: Minimum number of samples required to split an internal node (int or float, default=2)
  - If int, represents absolute number of samples
  - If float, represents a fraction of total samples

- `min_samples_leaf`: Minimum number of samples required to be at a leaf node (int or float, default=1)
  - Prevents creating too small leaves

- `max_features`: Number of features to consider when looking for the best split
  - None: all features
  - int: absolute number of features
  - float: fraction of features
  - 'sqrt': square root of total features
  - 'log2': log2 of total features

- `random_state`: Controls randomness in splitting

- `class_weight`: Weights associated with classes

3. Random Forest Hyperparameters:
- `n_estimators`: Number of trees in the forest (int, default=100)
  - More trees generally improve performance but increase computational cost

- `criterion`: Same as Decision Tree (default='gini')

- `max_depth`: Same as Decision Tree (default=None)

- `min_samples_split`: Same as Decision Tree (default=2)

- `min_samples_leaf`: Same as Decision Tree (default=1)

- `max_features`: Same as Decision Tree
  - Default is 'sqrt' for classification tasks

- `bootstrap`: Whether bootstrap samples are used when building trees (bool, default=True)
  - If False, the whole dataset is used to build each tree

- `random_state`: Controls randomness of bootstrapping and feature selection

- `n_jobs`: Number of jobs to run in parallel for both fitting and predicting

- `class_weight`: Weights for class balancing
