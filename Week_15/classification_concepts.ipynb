### Classification in Machine Learning

**Classification** is a supervised learning technique in machine learning used to categorize data into predefined classes or groups. It predicts discrete output labels based on input features. For example, classifying emails as *spam* or *not spam* or predicting whether a customer will churn.

---

### Key Concepts in Classification

1. **Classes/Labels**:
   - The predefined categories the model predicts.
   - Examples: *Spam/Not Spam*, *Disease/No Disease*.

2. **Features**:
   - Input variables (independent variables) used to make predictions.
   - Examples: Age, Income, Email Text.

3. **Training Dataset**:
   - Labeled data used to train the model.
   - Contains features and their corresponding true labels.

4. **Testing Dataset**:
   - Data used to evaluate the model's performance.

5. **Decision Boundary**:
   - A line (or curve) that separates classes in the feature space.

6. **Evaluation Metrics**:
   - Metrics to measure performance:
     - Accuracy
     - Precision
     - Recall
     - F1 Score
     - ROC-AUC

7. **Overfitting/Underfitting**:
   - Overfitting: Model performs well on training data but poorly on unseen data.
   - Underfitting: Model fails to capture patterns in the training data.

---

### Types of Classification

1. **Binary Classification**:
   - Predicts one of two classes.
   - Example: *Pass/Fail*, *True/False*.

2. **Multiclass Classification**:
   - Predicts one of three or more classes.
   - Example: Classifying handwritten digits (0–9).

3. **Multilabel Classification**:
   - Each instance can belong to multiple classes simultaneously.
   - Example: Tagging a movie with genres like *Action*, *Comedy*, and *Drama*.

4. **Imbalanced Classification**:
   - One class dominates the dataset compared to the others.
   - Example: Fraud detection (fraudulent transactions are rare).

5. **Hierarchical Classification**:
   - Classes are organized in a hierarchy.
   - Example: Categorizing animals into *Kingdom → Phylum → Class → Order*.

6. **Ordinal Classification**:
   - The output classes have a natural order.
   - Example: Predicting *Low/Medium/High* risk.
