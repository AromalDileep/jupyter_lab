# Diabetes predictions

```python
from sklearn.datasets import load_diabetes
```

```python
dataset = load_diabetes()
```

```python
dataset
```

```python
print(dataset['DESCR'])
```

```python
import pandas as pd 
df_diabetes = pd.DataFrame(dataset.data, columns= ['age',
  'sex',
  'bmi',
  'bp',
  's1',
  's2',
  's3',
  's4',
  's5',
  's6'])
df_diabetes.head()
```

```python
## Independent and dependenet features 
X = df_diabetes
y = dataset['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10 )
```

```python
X_train.head()
```

```python
## correlation (default = pearson correlation)
X_train.corr()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
plt.figure(figsize=(15,10))
sns.heatmap(X_train.corr(), annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train)
```

# Hyperparameter tuning

```python
params = {
    'criterion': ['squared error', 'friedman_mse', 'absoulte_error'],
    'splitter':['best', 'random'],
    'max_depth': [1,2,3,4,5,10,15,20,15],
    'max_features':['auto', 'sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
regressor = DecisionTreeRegressor()
```

```python
grid = GridSearchCV(regressor, param_grid= params, cv=5, scoring='neg_mean_squared_error')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
## predictions 
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
print('r2score: ', r2_score(y_test,y_pred))
print('mean absolute error: ', mean_absolute_error(y_test, y_pred))
print('mean squared error: ', mean_squared_error(y_test, y_pred))
```

```python
select_model = DecisionTreeRegressor(criterion= 'friedman_mse',
 max_depth= 2,
 max_features = 'log2',
 splitter ='best')
```
