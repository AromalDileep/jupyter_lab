```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns 
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv('/home/aromal/Documents/jupyter_Lab/Datasets_Main/Melbourne_housing_FULL.csv')
dataset.head()
```

```python
dataset.nunique() #unique values
```

```python
dataset.info()
```

```python
dataset.shape
```

```python
cols_to_use =['Suburb', 'Rooms', 'Type', 'Method', 'SellerG',
             'Regionname', 'Propertycount', 'Distance', 'CouncilArea',
             'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
dataset = dataset[cols_to_use]
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isna().sum()
```

```python
cols_to_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
dataset[cols_to_fill_zero] = dataset[cols_to_fill_zero].fillna(0)
```

```python
dataset.isna().sum()
```

```python
dataset['Landsize'] = dataset['Landsize'].fillna(dataset['Landsize'].mean())
# or 
# dataset['Landsize'].fillna(dataset['Landsize'].mean(), inplace=True)
dataset['BuildingArea'] = dataset['BuildingArea'].fillna(dataset['BuildingArea'].mean())
```

```python
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.isna().sum()
```

```python
dataset = pd.get_dummies(dataset, drop_first=True)
```

```python
dataset = dataset.astype(int)
```

```python
dataset
```

```python
X = dataset.drop('Price', axis=1)
y = dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
# r2_score(y_test, y_pred) # both are same
reg.score(X_test, y_test)
```

```python
reg.score(X_train, y_train)
```

```python
# so we can see it is clearly overfitting
# lasso regression L1
from sklearn.linear_model import Lasso

lasso_reg = Lasso(alpha=50, max_iter=100, tol=0.1)
lasso_reg.fit(X_train, y_train)
```

```python
lasso_reg.score(X_test, y_test)
```

```python
lasso_reg.score(X_train, y_train)
```

```python
# ridge regression L2
from sklearn.linear_model import Ridge 

ridge_reg = Ridge(alpha=50, max_iter=100, tol=0.1)
```

```python
ridge_reg.fit(X_train, y_train)
```

```python
ridge_reg.score(X_test, y_test)
```

```python
ridge_reg.score(X_train, y_train)
```
